```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
house_price_df = pd.read_csv("/content/drive/MyDrive/Kaggle Competition/train.csv")
house_price_test_df = pd.read_csv("/content/drive/MyDrive/Kaggle Competition/test.csv")
```

```python
house_price_df.head()
```

```python
house_price_test_df.head()
```

```python
house_price_df.describe()
```

```python

features = ["OverallQual","GarageArea" ,"GrLivArea","MSSubClass","LotFrontage"	,"LotArea","OverallQual"	,"OverallCond"	,"YearBuilt" ,	"YearRemodAdd"  ,	"TotalBsmtSF"	,"1stFlrSF"	,"2ndFlrSF"	]
y = house_price_df.SalePrice
X = house_price_df[features]
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
#
```

```python
house_price_model = RandomForestRegressor(random_state=1)
house_price_model.fit(train_X, train_y)
```

```python
val_predictions = house_price_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
X_test = house_price_test_df[["MoSold","YrSold", "GarageArea" , "BedroomAbvGr"	,"KitchenAbvGr","GrLivArea","MSSubClass","LotFrontage"	,"LotArea","OverallQual"	,"OverallCond"	,"YearBuilt" ,	"YearRemodAdd" ,"BsmtUnfSF",	"TotalBsmtSF"	,"1stFlrSF"	,"2ndFlrSF"	]]
prediction = house_price_model.predict(X_test)
```

```python
submission = pd.DataFrame({'Id': house_price_test_df.Id, 'SalePrice': prediction})
submission.to_csv('/content/drive/MyDrive/Kaggle Competition/submission.csv', index=False)
```
